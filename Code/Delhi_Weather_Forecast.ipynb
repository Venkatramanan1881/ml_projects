```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('/content/DailyDelhiClimateTrain.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
figure = px.line(data, x="date",
                 y="meantemp",
                 title='Mean Temperature in Delhi Over the Years')
figure.show()
```

```python
figure = px.line(data, x = 'date', y = 'humidity', title = "Mean Humidity over the years")
figure.show()
```

```python
figure = px.line(data, x = 'date', y = 'wind_speed', title = 'Windspeed over time')
figure.show()
```

```python
figure = px.scatter(data_frame = data, x = 'humidity', y = 'meantemp', size='meantemp', trendline = 'ols',
                    title = 'Relationship between Temperature and Humidity')
figure.show()
```

```python
data['date'] = pd.to_datetime(data['date'], format = '%m/%d/%Y')
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data.head()
```

```python
plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize = (15, 10))
plt.title('Temperature change in Delhi over Years')
sns.lineplot(data = data, x='month', y = 'meantemp', hue = 'year')
plt.show()
```

We can see that the temperature is rising every year in Delhi

```python
forecast_data = data.rename(columns = {'date': 'ds',
                                       'meantemp' : 'y'})
forecast_data
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
model = Prophet()
model.fit(forecast_data)
forecasts = model.make_future_dataframe(periods = 365)
predictions = model.predict(forecasts)
plot_plotly(model, predictions)
```
